```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
df = df.astype("int")
df.head()
```

```python
ip = df.drop(["Price"], axis="columns")
op = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ip, op, test_size=0.20, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Lasso() is L1 Regularization. For Very big dataset it perform very weel then normal linearRegression() model but, for small dataset things might be oposite.

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=1.0)
```

```python
lasso_reg.fit(X_train, y_train)
lasso_reg.score(X_test, y_test)
```

### Ridge() is L2 Regularization. Same as L1 Regularization but more improved

```python
from sklearn.linear_model import Ridge
Ridge_reg = Ridge(alpha=50, max_iter=100, tol=1.0)
```

```python
Ridge_reg.fit(X_train, y_train)
Ridge_reg.score(X_test, y_test)
```
