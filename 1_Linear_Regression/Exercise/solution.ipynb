```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 6))
plt.title("Canada per capital income")
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(x=df.year, y=df["per capita income (US$)"], color="blue", marker="x")
plt.show()
```

```python
lreg = linear_model.LinearRegression()
lreg.fit(df[["year"]], df[["per capita income (US$)"]])
```

```python
lreg.predict([[2020]])
```

```python
plt.figure(figsize=(12, 6))
plt.title("Canada per capital income")
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(x=df.year, y=df["per capita income (US$)"], color="blue", marker="x")
plt.plot(df.year, lreg.predict(df[["year"]]), color="black")
plt.show()
```
