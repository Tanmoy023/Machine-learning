```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprice.csv")
df
```

# Ploting our inputs csv


```python
plt.scatter(x=df.area, y=df.price, color="green", marker="x")

plt.title("Price of a house according to area")
plt.xlabel("Area(sqr-ft)")
plt.ylabel("Price($)")

plt.show()
```

# Creating a Linear Regression object


```python
L_reg = linear_model.LinearRegression()
L_reg.fit(df[["area"]], df.price)  # train this model
```

# Predict


```python
# L_reg.predict([[3300]])
L_reg.predict([[3300],[5000]])
# L_reg.predict([[3300],[4000],[5000],[6000]])
```

```python
m = L_reg.coef_  # coeficient
b = L_reg.intercept_  # intercept
y = m * 3300 + b
y
```

```python
areas = pd.read_csv("areas.csv")
areas.head()
```

```python
predictions = L_reg.predict(areas)
predictions
```

```python
areas["predicted price"] = predictions
areas
```

```python
areas.to_csv("Prediction.csv", index=False)
```

```python
plt.scatter(x=df.area, y=df.price, color="red", marker="x")

plt.title("Price of a house according to area", fontsize=20)
plt.xlabel("Area(sqr-ft)", fontsize=15)
plt.ylabel("Price($)", fontsize=15)

plt.plot(df.area, L_reg.predict(df[["area"]]), color="green")
```
