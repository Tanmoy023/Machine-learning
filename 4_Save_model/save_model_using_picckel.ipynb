```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprice.csv")
df
```

```python
plt.title("Price of a house according to area")
plt.xlabel("Area(sqr-ft)")
plt.ylabel("Price($)")
plt.scatter(x=df.area, y=df.price, color="green", marker="x")
```

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(df[["area"]], df.price)  # train this model
```

```python
lr_model.predict([[3300]])
# lr_model.predict([[3300],[5000]])
# lr_model.predict([[3300],[4000],[5000],[6000]])
```

```python
import pickle

with open('model_pickle','wb') as f:
    pickle.dump(lr_model,f)
```

```python
with open('model_pickle','rb') as r:
    model = pickle.load(r)
```

```python
model.predict([[5000]])
```
