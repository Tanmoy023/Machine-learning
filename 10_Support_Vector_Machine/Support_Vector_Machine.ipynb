```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
from matplotlib import pyplot as plt
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('satosa and versicolor classification')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker="x")
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker="o")
plt.show()
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('satosa and versicolor classification')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker="x")
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', marker="o")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower_name'],axis='columns')
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

```python
from sklearn.svm import SVC
model = SVC()
```
